#### Câu 1: (2 điểm)
Viết chương trình tạo và vẽ tín hiệu x gồm các yêu cầu sau: <br>
> x là tổng hợp của 3 tín hiệu hình sin: <br>
>> a1 = 2, f1 = 8Hz, pha = 0 <br>
>> a2 = 3, f2 = 11 Hz, pha = $\pi/6$ <br>
>> a3 = 0.8, f3 = 42 Hz, pha3 = $\pi/3$ <br>

> Tần số lấy mẫu: fs = 1 KHz<br>
> Thời gian vẽ: 2 s


```python
import numpy as np
import matplotlib.pyplot as plt

fs=1000
# Tạo mảng thời gian t
t = np.linspace(0, 2, fs * 2)

# Tính toán các tín hiệu hình sin
y1 = 2 * np.sin(2 * np.pi * 8 * t + 0)
y2 = 3 * np.sin(2 * np.pi * 11 * t + np.pi / 6)
y3 = 0.8 * np.sin(2 * np.pi * 42 * t + np.pi / 3)

# Tính toán tín hiệu tổng hợp
y = y1 + y2 + y3

# Vẽ tín hiệu lên đồ thị
plt.plot(t, y)
plt.xlabel('Thời gian (s)')
plt.ylabel('Điện áp (V)')
plt.title('Tín hiệu x')
plt.show()
```
